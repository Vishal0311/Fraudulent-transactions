## Feature engg

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder={}
for i in df.select_dtypes('object').columns:
    encoder[i]=LabelEncoder()
    df[i]=encoder[i].fit_transform(df[i])
```

```python
df.head()
```

```python
x=df.drop(columns=['isFraud'])
y=df['isFraud']
```

```python
y.value_counts()
```

```python
x.head()
```

```python
pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
over_sample=SMOTE(random_state=0)
x,y=over_sample.fit_resample(x,y)
```

```python
y.value_counts()
```

## Feature scalling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
```

## Train-test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

## Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
y_pred= log_reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

### as of now we have a accuracy score of 92% which is not so good.

### let us now remove multicollinearity and selecting best feature

```python
from sklearn.model_selection import cross_val_score
```

```python
cvs = cross_val_score(log_reg,x,y,cv=3)
print(cvs)
```

```python
cvs.mean()
```

#### as we can see that cross val score is less than prediction score therefore this is the case of underfitting

```python
from sklearn.feature_selection import chi2, SelectKBest
```

```python
best_feature = SelectKBest(chi2,k=8)
kbest=best_feature.fit_transform(x,y)
np.array(df.drop(columns=['isFraud']).columns)[best_feature.get_support()]
```

```python
pd.DataFrame(list(zip(df.drop(columns=['isFraud']).columns, best_feature.scores_)),columns=['features','scores']).sort_values(by='scores',ascending=False)
```
